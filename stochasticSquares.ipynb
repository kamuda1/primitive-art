```python
im = Image.new('RGB', (1000, 1000))
for i in np.linspace(500, 5, 50):
    for _ in range(50):
        make_squares(im, i, i, color=(int(256 % (10*np.cos(i))),
                                      int(256 % (20*np.sin(i))),
                                      int(i % 256)))
display(im)
```

```python
im = Image.new('RGB', (1000, 1000))
for i in np.linspace(500, 5, 50):
    for _ in range(50):
        height_mag = 75
        width_mag = 2
        height = np.random.uniform(-height_mag, height_mag)
        width = np.random.uniform(-width_mag, width_mag)
        make_squares(im,height,width,color=(int(556 % i),
                                            int(556 % i),
                                            int(556 % i)))
display(im)
```

```python
im = Image.new('RGB', (1366, 768))
for i in np.linspace(500, 5, 50):
    for _ in range(50):
        
        height_mag = im.size[0]/10
        width_mag = im.size[1]/10
        
        [x,y] = [np.random.uniform(im.size[0]),
                 np.random.uniform(im.size[1])]
        height = np.random.uniform(-height_mag, height_mag)
        width = np.random.uniform(-width_mag, width_mag)
        make_squares(im, height, width, color=(int(20/(im.size[0]/10.)*x+556 % i),
                                               int(556 % i),
                                               int(-40/(im.size[0]/10.)*x+556 % i)))

display(im)
```
