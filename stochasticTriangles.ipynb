```python
from PIL import Image, ImageDraw
import PIL
from IPython.display import display
import numpy as np
from PIL import ImageFilter
from numpy import cos, sin
```

```python
from primatives import make_triangle
```

```python
im = Image.new('RGB', (1000, 1000))
for i in range(5, 50):
    for _ in range(50):
        make_triangle(im, i, i, color=(256 % i,
                                       256 % i,
                                       256 % i))
display(im)
```

```python
im = Image.new('RGB', (1000, 1000))
for i in np.linspace(500, 5, 50):
    for _ in range(50):
        make_triangle(im, i, i, color=(int(556 % i),
                                       int(556 % i),
                                       int(556 % i)))
display(im)
```

```python
im = Image.new('RGB', (1000, 1000))
for i in np.linspace(500, 5, 50):
    for _ in range(50):
        make_triangle(im, i, i, color=(int(i % 256),
                                       int(i % 256),
                                       int(i % 256)))
display(im)
```

```python
im = Image.new('RGB', (1000, 1000))
for i in np.linspace(500, 5, 50):
    for _ in range(50):
        make_triangle(im, i, i, color=(int(256 % (10*np.cos(i))),
                                       int(256 % (20*np.sin(i))),
                                       int(i % 256)))
display(im)
```

```python
im = Image.new('RGB', (1000, 1000))
for i in np.linspace(500, 5, 50):
    for _ in range(50):

        height_mag = 75
        width_mag = 2

        [x, y] = [np.random.uniform(im.size[0]),
                  np.random.uniform(im.size[1])]
        height = np.random.uniform(-height_mag, height_mag)
        width = np.random.uniform(-width_mag, width_mag)
        make_triangle(im, height, width, color=(int(556 % i),
                                                int(556 % i),
                                                int(556 % i)))

display(im)
```

```python
im = Image.new('RGB', (1000, 1000))
for i in np.linspace(500,5,50):
    for _ in range(50):
        
        height_mag = 100
        width_mag = 80
        
        [x,y] = [np.random.uniform(im.size[0]),
                 np.random.uniform(im.size[1])]
        height = np.random.uniform(-height_mag, height_mag)
        width = np.random.uniform(-width_mag, width_mag)

        draw = ImageDraw.Draw(im)
        draw.polygon(((x, y),
                      (x, y+height),
                      (x+width, y),
                      (x, y)),
                     fill=(int(0.20*x+556 % i),
                           int(556 % i),
                           int(-0.40*x+556 % i)))
display(im)
```

```python
im = Image.new('RGB', (2*1366, 2*768))
for i in np.linspace(500, 5, 50):
    for _ in range(300):
        
                
        height_mag = im.size[0]/13.
        width_mag = im.size[1]/400.
        
        [x,y] = [np.random.uniform(im.size[0]),
                 np.random.uniform(im.size[1])]
        height = np.random.uniform(-height_mag, height_mag)
        width = np.random.uniform(-width_mag, width_mag)

        draw = ImageDraw.Draw(im)
        theta = 0.0001
        draw.polygon(((x*cos(x*theta), y),
                      (x, y+height),
                      (x+width, y),
                      (x, y)),
                     fill=(int(556 % i*x),
                           int(556 % i*x/300),
                           int(556 % i)))

im.filter(ImageFilter.SMOOTH_MORE)
```
